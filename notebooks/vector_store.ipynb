```python
from langchain.text_splitter import CharacterTextSplitter #use this to split all text for embedding
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders.csv_loader import CSVLoader
from langchain_huggingface import HuggingFaceEmbeddings


from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
class VectorStoreBuilder:
    def __init__(self, csv_path:str, persist_dir:str="chroma_db"): #creates chroma_db dir where vecotrstore will be created
        self.csv_path = csv_path
        self.persis_dir = self.persis_dir
        self.embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2") #model to create embeddings post splitting

    def build_and_save_vectorestore(self):
        #create loader
        loader = CSVLoader(
            file_path=self.csv_path,
            encoding='utf-8',
            metadata_columns=[]
        )
        #load date
        data = loader.load()
        #split
        splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
        texts = splitter.split_documents(data)

        #convert into embeddings and stor
        db = Chroma.from_documents(texts, self.embeddings, persist_directory=self.persis_dir)
        db.persist()

    def load_vectorstore(self):
        return Chroma(persist_directory=self.persis_dir, embedding_function=self.embedding)

    
```
