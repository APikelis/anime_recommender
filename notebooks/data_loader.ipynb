```python
class AnimeDataLaoder:
    def __init__(self, original_csv: str, processed_csv:str): #csvs = paths of files
        self.original_csv = original_csv
        self.processed_csv = processed_csv

    def load_and_process(self):
        df = pd.read_csv(self.original_csv, encoding ='utf-8', error_bad_lines=False).dropna()#convert and read in utf-8 encoding (standard)
        required_cols = {'Name', 'Genre', 'synopsis'} #dictioary of cols
        missing = required_cols - set(df.columns) #checks missing columns
        if missing:
            raise ValueError("Missing Column in CSV File")
        
        #create new column that combines all relevant data for our LLM to read
        df['combined_info'] = (
            "Title: " + df['Name'] + ".. Overview: " + df["synopsis"] + "Genres: " + df["Genres"]
        )

        #save to new csv with only combined_info column
        df[['combined_info']].to_csv(self.processed_csv, index=False, encoding='utf-8')

        return self.processed_csv
    
```

```python
    #self lets us refer to the current an instance of the class

a = AnimeDataLaoder("file1.csv", "out1.csv")
b = AnimeDataLaoder("file2.csv", "out2.csv")

#both use same class but self keeps data separate
```
